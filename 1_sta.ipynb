1. What is a random variable in probability theory?

A random variable is a fundamental concept in probability theory. It is a function that assigns a real number to each outcome in the sample space of a random experiment. In simpler terms, it translates outcomes of random processes into numerical values, which makes it easier to analyze, quantify, and compute probabilities.

Random variables are of two primary types: discrete (which take a countable number of values) and continuous (which take values from an uncountably infinite set, such as real numbers within an interval).

For example, when rolling a six-sided die, the number that appears is a discrete random variable that can take values from 1 to 6. If we were measuring the amount of rainfall in a day, this would be a continuous random variable since rainfall can be any non-negative real number.

2. What are the types of random variables?

There are two main types of random variables:

Discrete Random Variables: These take on a finite or countably infinite set of distinct values. Examples include the number of children in a family, the number of cars passing a checkpoint, or the outcome of rolling dice.

Continuous Random Variables: These can take on any value within a given range or interval. They are uncountably infinite and typically represent measurements such as height, weight, time, or temperature.

Each type is associated with a different type of probability distribution and requires distinct mathematical tools for analysis.

3. What is the difference between discrete and continuous distributions?

The distinction lies in the type of values the random variable can assume:

Discrete Distributions describe the probability of outcomes of a discrete random variable. Probabilities are assigned to specific individual values. The sum of these probabilities equals 1. Common examples include the binomial and Poisson distributions.

Continuous Distributions describe outcomes of continuous random variables. Instead of individual probabilities, they use probability density functions (PDFs) to describe the likelihood of values within an interval. Probabilities are computed over intervals (e.g., the probability a value lies between 2 and 4), and the area under the PDF curve over that interval represents the probability.

4. What are probability distribution functions (PDF)?

The term Probability Distribution Function (PDF) generally refers to two related concepts depending on whether the variable is discrete or continuous:

For discrete variables, it’s called a probability mass function (PMF). It gives the probability that a discrete random variable is exactly equal to some value.

For continuous variables, the PDF is a function that describes the relative likelihood for a random variable to take on a given value. The probability that the variable falls within a particular interval is the area under the curve of the PDF over that interval.

In both cases, the distribution function fully characterizes the behavior of the random variable in question.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

The Central Limit Theorem states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the original distribution, provided the samples are independent and identically distributed (i.i.d.) with finite mean and variance.

Importance:

Justifies the use of the normal distribution for inference

Underpins hypothesis testing, confidence intervals, and many statistical methods

23. How does a probability distribution relate to the expected outcome of a random variable?

A probability distribution provides the framework for determining the expected outcome of a random variable. The expected value is essentially a weighted average of all possible outcomes, with the weights being their probabilities.

The shape of the distribution influences the likelihood of certain outcomes.

The mean (expected value) summarizes the central tendency of the distribution.

For example, a skewed distribution might have a mean that doesn't coincide with the most probable outcome (mode), but it still accurately reflects the average result over many trials.

```python
#9  Write a Python function to calculate Z-scores from a dataset and plot them>
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def plot_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std

    sns.histplot(z_scores, bins=30, kde=True)
    plt.title("Z-scores of Dataset")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.axvline(0, color='red', linestyle='--', label='Mean')
    plt.legend()
    plt.show()
    return z_scores

# Example usage
if __name__ == "__main__":
    sample_data = np.random.normal(0, 1, 1000)
    plot_z_scores(sample_data)
```

```python
#10 : Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def simulate_clt(dist_func, sample_size=30, n_samples=1000):
    means = [np.mean(dist_func(size=sample_size)) for _ in range(n_samples)]
    sns.histplot(means, kde=True, bins=30)
    plt.title(f"CLT Simulation (Sample Size={sample_size}, n={n_samples})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")
    plt.show()

# Example using exponential distribution
if __name__ == "__main__":
    simulate_clt(lambda size: np.random.exponential(scale=2, size=size))
```

```python
#11  Simulate multiple samples from a normal distribution and verify the Central Limit Theorem5
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def verify_clt(sample_size=30, n_samples=1000, mu=10, sigma=5):
    sample_means = []

    for _ in range(n_samples):
        sample = np.random.normal(loc=mu, scale=sigma, size=sample_size)
        sample_means.append(np.mean(sample))

    # Plot the distribution of sample means
    plt.figure(figsize=(8, 5))
    sns.histplot(sample_means, bins=30, kde=True, color='skyblue', edgecolor='black')
    plt.title(f"Central Limit Theorem: Means of {n_samples} samples (n={sample_size})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")
    plt.axvline(mu, color='red', linestyle='--', label='True Mean')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
verify_clt()
```

```python
#14  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
#distribution5
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    z = (x - mean) / std_dev
    prob = norm.cdf(z)
    print(f"Z-score: {z:.2f}")
    print(f"Probability that a standard normal variable is <= {x}: {prob:.4f}")
    return z, prob

# Example usage
calculate_z_score(x=85, mean=70, std_dev=10)
```

```python
#15  Implement hypothesis testing using Z-statistics for a sample dataset5
from scipy.stats import norm
import numpy as np

def z_test(sample_mean, population_mean, population_std, n, alpha=0.05):
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Z-statistic: {z:.4f}")
    print(f"P-value: {p_value:.4f}")
    if p_value < alpha:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")

# Example usage
z_test(sample_mean=105, population_mean=100, population_std=15, n=36)
```

```python
#16  Create a confidence interval for a dataset using Python and interpret the result5
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    z = norm.ppf(1 - (1 - confidence) / 2)
    margin = z * std_err
    lower = mean - margin
    upper = mean + margin

    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    return lower, upper

# Example usage
data = np.random.normal(100, 15, size=40)
confidence_interval(data)
```

```python
#17  Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean5
import numpy as np
from scipy.stats import norm

def confidence_interval_normal(mu=50, sigma=10, n=50, confidence=0.95):
    # Generate normal data
    data = np.random.normal(mu, sigma, n)

    # Compute sample mean and standard error
    sample_mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)

    # Z-value for the desired confidence level
    z = norm.ppf(1 - (1 - confidence) / 2)

    # Confidence interval
    lower = sample_mean - z * std_err
    upper = sample_mean + z * std_err

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print("Interpretation: We're", int(confidence*100), "% confident the true population mean lies in this interval.")

# Run the function
confidence_interval_normal()
```

```python
#18  Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution5
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean=0, std=1):
    x = np.linspace(mean - 4*std, mean + 4*std, 1000)
    y = norm.pdf(x, mean, std)

    plt.plot(x, y, label=f'N({mean}, {std}²)', color='green')
    plt.title("Probability Density Function of Normal Distribution")
    plt.xlabel("x")
    plt.ylabel("PDF")
    plt.grid(True)
    plt.legend()
    plt.show()

# Run the function
plot_normal_pdf()
```

```python
#19  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution5
from scipy.stats import poisson

def poisson_cdf(lambda_val=3, x=5):
    cdf = poisson.cdf(x, mu=lambda_val)
    print(f"P(X ≤ {x}) for Poisson(λ={lambda_val}) = {cdf:.4f}")
    print("Interpretation: There's a", round(cdf*100, 2), "% chance of observing", x, "or fewer events.")

# Run the function
poisson_cdf()
```

```python
#20 Simulate a random variable using a continuous uniform distribution and calculate its expected value5
import numpy as np

def simulate_uniform(a=10, b=20, n=1000):
    data = np.random.uniform(a, b, n)
    expected_value = np.mean(data)
    theoretical = (a + b) / 2
    print(f"Simulated Expected Value: {expected_value:.2f}")
    print(f"Theoretical Expected Value: {theoretical:.2f}")

# Run the function
simulate_uniform()
```

```python
#21  Write a Python program to compare the standard deviations of two datasets and visualize the difference5
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def compare_std():
    data1 = np.random.normal(50, 5, 1000)
    data2 = np.random.normal(50, 15, 1000)

    std1 = np.std(data1)
    std2 = np.std(data2)

    print(f"Standard Deviation of Dataset 1: {std1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std2:.2f}")

    plt.figure(figsize=(8, 5))
    sns.kdeplot(data1, label='SD = 5', fill=True)
    sns.kdeplot(data2, label='SD = 15', fill=True)
    plt.title("Comparison of Standard Deviations")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the function
compare_std()
```

```python
# 22  Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution5
import numpy as np

def range_and_iqr():
    data = np.random.normal(100, 20, 1000)
    data_range = np.ptp(data)
    q75, q25 = np.percentile(data, [75, 25])
    iqr = q75 - q25

    print(f"Range: {data_range:.2f}")
    print(f"IQR (Interquartile Range): {iqr:.2f}")

# Run the function
range_and_iqr()
```

```python
#23 Implement Z-score normalization on a dataset and visualize its transformation5
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def z_score_normalization():
    data = np.random.normal(60, 15, 1000)
    z_scores = (data - np.mean(data)) / np.std(data)

    plt.figure(figsize=(8, 5))
    sns.histplot(data, bins=30, kde=True, label='Original Data', color='skyblue')
    sns.histplot(z_scores, bins=30, kde=True, label='Z-score Normalized', color='orange')
    plt.title("Z-score Normalization Comparison")
    plt.xlabel("Value / Z-score")
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the function
z_score_normalization()
```

```python
#24  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
# distribution
import numpy as np
from scipy.stats import skew, kurtosis

def skew_kurtosis():
    data = np.random.normal(0, 1, 1000)
    skewness = skew(data)
    kurt = kurtosis(data)  # Excess kurtosis

    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (Excess): {kurt:.4f}")
    print("Interpretation:")
    print("- Skewness ~0 indicates symmetry.")
    print("- Excess kurtosis ~0 indicates a normal peak.")

# Run the function
skew_kurtosis()

```
